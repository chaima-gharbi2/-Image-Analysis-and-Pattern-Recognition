```python
T = [[0,1], [2,3]]
U = [[4,5], [6,7]]

print(T + U)  # List concatenation
print(2 * T)  # List replication
```

nterpretation:

T + U concatenates the lists instead of performing mathematical addition

# **Convert the previous lists into NumPy arrays and interpret the results.**

```python
import numpy as np

T2 = np.array(T)
U2 = np.array(U)

print(T2 + U2)  # Element-wise addition
print(2 * T2)   # Element-wise multiplication
```

 Conclusion: Unlike Python lists, NumPy supports mathematical operations directly on arrays.

```python
import numpy as np

T3 = np.array([[0, 1], [2, 3], [4, 5]])
U3 = np.array([[4, 5], [6, 7], [6, 5]])

print(T3 + U3)  # Element-wise addition
```

# **Display specific elements of array T3**

```python
# Display the 2nd row
print(T3[1])

# Display the 1st column
print(T3[:, 0])

# Display the element at row 2, column 1
print(T3[1, 0])

```

# **Modify an element in the array (T3[2,0] = 10).**

```python
T3[2, 0] = 10
print(T3)
```

Conclusion: NumPy allows direct modification of individual elements in an array.

# **Copy an array using two methods and interpret the result.**

```python
T2 = T  # Copy by reference
T3 = np.copy(T)  # Independent copy

T[0][0] = 6  # Modify T

print(T)   # Modified
print(T2)  # Modified (because T2 is a reference to T)
print(T3)  # Unchanged (independent copy)
```

Interpretation:

T2 = T only creates a reference, so modifying T also modifies T2.
T3 = np.copy(T) creates an independent copy, so T3 remains unchanged.

# ***Image Processing with Python***

# Read and display an image ("Lena.jpg").

```python
from PIL import Image
import numpy as np

image = Image.open("/content/Lena (1).jpg")
image.show()  # Display the image
```

# Display image properties.

```python
# Image size (width, height)
print(image.size)

# Number of rows (height)
print(image.size[1])

# Number of columns (width)
print(image.size[0])

# Number of color components
image_array = np.array(image)
print(image_array.shape[-1])
```

#Extract and save only the red component of the image

```python
image_array = np.array(image)
red_channel = image_array.copy()
red_channel[:, :, 1] = 0  # Set green component to 0
red_channel[:, :, 2] = 0  # Set blue component to 0

red_image = Image.fromarray(red_channel)
red_image.save("Lena_Red.jpg")
red_image.show()
```

# Convert the image to grayscale and save it

```python
gray_image = image.convert("L")
gray_image.save("Lena_Gray.jpg")
gray_image.show()
```

```python
import matplotlib.pyplot as plt
plt.imshow(gray_image, cmap='gray')
plt.axis('off')  # Optional: to hide the axes
plt.show()
```
